```python
# Q. trip_distance의 이상치를 확인합니다.

# trip_distance 열의 이상치를 확인하기 위해 상자 그림을 그립니다.
plt.figure(figsize=(8, 6))
sns.boxplot(x=data['trip_distance'])
plt.title('Box plot of Trip Distance')
plt.xlabel('Trip Distance')
plt.show()
```

```python
# Define a threshold for trip_distance based on the box plot visualization
# (e.g., values significantly higher than the upper whisker)
# Based on the previous box plot, a threshold around 10-15 seems reasonable to start.
# Let's choose 15 as a starting point and we can adjust it if needed.
threshold = 15

# Remove rows where 'trip_distance' is greater than the threshold
data_cleaned = data[data['trip_distance'] <= threshold].copy()

# Display the number of rows before and after removing outliers
print(f"Number of rows before removing outliers: {len(data)}")
print(f"Number of rows after removing outliers: {len(data_cleaned)}")

# Display the box plot of the cleaned data to verify outlier removal
plt.figure(figsize=(8, 6))
sns.boxplot(x=data_cleaned['trip_distance'])
plt.title('Box plot of Trip Distance after Outlier Removal (Threshold = 15)')
plt.xlabel('Trip Distance')
plt.show()
```

```python
# trip_distance 열의 히스토그램을 그립니다.
plt.figure(figsize=(10, 6))
sns.histplot(data_cleaned['trip_distance'], bins=20, kde=True)
plt.title('Histogram of Trip Distance')
plt.xlabel('Trip Distance')
plt.ylabel('Frequency')
plt.show()
```

```python
# Recreate data_cleaned from the original data after all feature engineering steps
data_cleaned = data.copy()

# Remove outliers in 'trip_distance' based on the previously determined threshold (15)
threshold_trip_distance = 15
data_cleaned = data_cleaned[data_cleaned['trip_distance'] <= threshold_trip_distance].copy()

# Remove outliers in 'fare_amount' where fare_amount is <= 0
data_cleaned = data_cleaned[data_cleaned['fare_amount'] > 0].copy()

# Define a threshold for travel_time_seconds based on the scatter plot visualization
threshold_travel_time = 5000

# Remove rows where 'travel_time_seconds' is greater than the threshold
data_cleaned = data_cleaned[data_cleaned['travel_time_seconds'] <= threshold_travel_time].copy()

# Display the number of rows after removing all identified outliers
print(f"Number of rows after removing all identified outliers: {len(data_cleaned)}")

# Display the scatter plot of trip_distance vs travel_time_seconds for the cleaned data
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data_cleaned['travel_time_seconds'], y=data_cleaned['trip_distance'])
plt.title('Scatter plot of Trip Distance vs Travel Time (Seconds) after Outlier Removal')
plt.xlabel('Travel Time (Seconds)')
plt.ylabel('Trip Distance')
plt.show()

# Display the scatter plot of fare_amount vs travel_time_seconds for the cleaned data
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data_cleaned['travel_time_seconds'], y=data_cleaned['fare_amount'])
plt.title('Scatter plot of Fare Amount vs Travel Time (Seconds) after Outlier Removal')
plt.xlabel('Travel Time (Seconds)')
plt.ylabel('Fare Amount')
plt.show()
```
