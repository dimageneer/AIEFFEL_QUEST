## **프로젝트: 새로운 데이터셋으로 나만의 이미지 분류기 만들어보기**

```python
import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
import numpy as np

# ==========================================
# Step 1. 데이터셋 내려받기 & 나누기
# ==========================================
# tf_flowers는 기본적으로 'train' 스플릿 하나만 있습니다.
# 이를 80% 학습, 10% 검증, 10% 테스트로 직접 나눕니다.
(train_ds, val_ds, test_ds), metadata = tfds.load(
    'tf_flowers',
    split=['train[:80%]', 'train[80%:90%]', 'train[90%:]'],
    with_info=True,
    as_supervised=True,
)

print(f"클래스 개수: {metadata.features['label'].num_classes}")
print(f"클래스 이름: {metadata.features['label'].names}")
# 출력: ['dandelion', 'daisy', 'tulips', 'sunflowers', 'roses'] (순서는 다를 수 있음)

# ==========================================
# Step 2. 데이터셋 전처리 (Formatting)
# ==========================================
IMG_SIZE = 224  # VGG16 입력 크기

def format_image(image, label):
    image = tf.image.resize(image, (IMG_SIZE, IMG_SIZE))
    image = image / 255.0  # 0~1 사이로 정규화
    return image, label

BATCH_SIZE = 32

def prepare_data(ds, shuffle=False):
    ds = ds.map(format_image)
    if shuffle:
        ds = ds.shuffle(1000)
    ds = ds.batch(BATCH_SIZE)
    ds = ds.prefetch(tf.data.experimental.AUTOTUNE)
    return ds

train_batches = prepare_data(train_ds, shuffle=True)
validation_batches = prepare_data(val_ds)
test_batches = prepare_data(test_ds)

# ==========================================
# Step 3. 모델 설계하기 (VGG16 기반 + Custom Head)
# ==========================================
base_model = tf.keras.applications.VGG16(
    input_shape=(IMG_SIZE, IMG_SIZE, 3),
    include_top=False,
    weights='imagenet'
)
base_model.trainable = False  # Base Model 고정 (Feature Extractor로 사용)

model = tf.keras.Sequential([
    base_model,
    tf.keras.layers.GlobalAveragePooling2D(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.BatchNormalization(), # 추가된 정규화 기법
    tf.keras.layers.Dropout(0.5),         # 과적합 방지를 위한 Dropout
    # 중요: 클래스가 5개이므로 출력 노드는 5, 활성화 함수는 softmax
    tf.keras.layers.Dense(5, activation='softmax')
])

model.summary()

# ==========================================
# Step 4. 모델 학습시키기
# ==========================================
# Optimizer를 Adam으로, learning_rate를 조금 낮게 설정해 봅니다.
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
              loss='sparse_categorical_crossentropy', # 정수 레이블(0,1,2..)일 때 사용
              metrics=['accuracy'])

# 학습 조기 종료(Early Stopping) 콜백 추가
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss', patience=3, restore_best_weights=True
)

history = model.fit(
    train_batches,
    epochs=15, # Epoch 늘리기
    validation_data=validation_batches,
    callbacks=[early_stopping]
)

# ==========================================
# Step 5. 모델 성능 평가하기
# ==========================================
# 1. 학습 과정 그래프
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()

# 2. 테스트 셋 평가
loss, accuracy = model.evaluate(test_batches)
print(f'\nTest set accuracy: {accuracy*100:.2f}%')

# ==========================================
# Step 6. 모델 활용하기 (예측)
# ==========================================
# 테스트 데이터셋에서 이미지를 하나 가져와 예측해 봅니다.
image_batch, label_batch = next(iter(test_batches))
image = image_batch[0]
true_label = label_batch[0]

# 모델 예측
predictions = model.predict(tf.expand_dims(image, 0))
predicted_class = np.argmax(predictions)
class_names = metadata.features['label'].names

plt.figure()
plt.imshow(image)
plt.title(f"True: {class_names[true_label]}, Pred: {class_names[predicted_class]}")
plt.axis('off')
plt.show()
```

## 데이터 증강 레이어 추가

### Subtask:
모델의 입력단에 `tf.keras.layers.RandomFlip`, `RandomRotation`, `RandomZoom`과 같은 데이터 증강 레이어들을 추가하여 데이터의 다양성을 높이고 과적합을 방지합니다.


**Reasoning**:
To implement data augmentation, I will redefine the model to include `RandomFlip`, `RandomRotation`, and `RandomZoom` layers at the input of the `Sequential` model, before the pre-trained `VGG16 base_model`, while retaining the subsequent custom head layers and ensuring the `base_model` remains frozen.



```python
import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
import numpy as np

# ==========================================
# Step 1. 데이터셋 내려받기 & 나누기
# ==========================================
# tf_flowers는 기본적으로 'train' 스플릿 하나만 있습니다.
# 이를 80% 학습, 10% 검증, 10% 테스트로 직접 나눕니다.
(train_ds, val_ds, test_ds), metadata = tfds.load(
    'tf_flowers',
    split=['train[:80%]', 'train[80%:90%]', 'train[90%:]'],
    with_info=True,
    as_supervised=True,
)

print(f"클래스 개수: {metadata.features['label'].num_classes}")
print(f"클래스 이름: {metadata.features['label'].names}")
# 출력: ['dandelion', 'daisy', 'tulips', 'sunflowers', 'roses'] (순서는 다를 수 있음)

# ==========================================
# Step 2. 데이터셋 전처리 (Formatting)
# ==========================================
IMG_SIZE = 224  # VGG16 입력 크기

def format_image(image, label):
    image = tf.image.resize(image, (IMG_SIZE, IMG_SIZE))
    image = image / 255.0  # 0~1 사이로 정규화
    return image, label

BATCH_SIZE = 32

def prepare_data(ds, shuffle=False):
    ds = ds.map(format_image)
    if shuffle:
        ds = ds.shuffle(1000)
    ds = ds.batch(BATCH_SIZE)
    ds = ds.prefetch(tf.data.experimental.AUTOTUNE)
    return ds

train_batches = prepare_data(train_ds, shuffle=True)
validation_batches = prepare_data(val_ds)
test_batches = prepare_data(test_ds)

# ==========================================
# Step 3. 모델 설계하기 (VGG16 기반 + Custom Head) - 데이터 증강 레이어 추가
# ==========================================
base_model = tf.keras.applications.VGG16(
    input_shape=(IMG_SIZE, IMG_SIZE, 3),
    include_top=False,
    weights='imagenet'
)
base_model.trainable = False  # Base Model 고정 (Feature Extractor로 사용)

# 데이터 증강 레이어 추가
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip('horizontal_and_vertical'),
    tf.keras.layers.RandomRotation(0.2),
    tf.keras.layers.RandomZoom(0.2),
], name='data_augmentation')

model = tf.keras.Sequential([
    data_augmentation,          # 데이터 증강 레이어
    base_model,
    tf.keras.layers.GlobalAveragePooling2D(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.BatchNormalization(), # 추가된 정규화 기법
    tf.keras.layers.Dropout(0.5),         # 과적합 방지를 위한 Dropout
    # 중요: 클래스가 5개이므로 출력 노드는 5, 활성화 함수는 softmax
    tf.keras.layers.Dense(5, activation='softmax')
])

model.summary()

# ==========================================
# Step 4. 모델 학습시키기
# ==========================================
# Optimizer를 Adam으로, learning_rate를 조금 낮게 설정해 봅니다.
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
              loss='sparse_categorical_crossentropy', # 정수 레이블(0,1,2..)일 때 사용
              metrics=['accuracy'])

# 학습 조기 종료(Early Stopping) 콜백 추가
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss', patience=3, restore_best_weights=True
)

history = model.fit(
    train_batches,
    epochs=15, # Epoch 늘리기
    validation_data=validation_batches,
    callbacks=[early_stopping]
)

# ==========================================
# Step 5. 모델 성능 평가하기
# ==========================================
# 1. 학습 과정 그래프
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()

# 2. 테스트 셋 평가
loss, accuracy = model.evaluate(test_batches)
print(f'\nTest set accuracy: {accuracy*100:.2f}%')

# ==========================================
# Step 6. 모델 활용하기 (예측)
# ==========================================
# 테스트 데이터셋에서 이미지를 하나 가져와 예측해 봅니다.
image_batch, label_batch = next(iter(test_batches))
image = image_batch[0]
true_label = label_batch[0]

# 모델 예측
predictions = model.predict(tf.expand_dims(image, 0))
predicted_class = np.argmax(predictions)
class_names = metadata.features['label'].names

plt.figure()
plt.imshow(image)
plt.title(f"True: {class_names[true_label]}, Pred: {class_names[predicted_class]}")
plt.axis('off')
plt.show()
```
